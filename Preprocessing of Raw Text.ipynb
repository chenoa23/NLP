```python
# Load the text corupus to a variable
with open('/content/drive/MyDrive/CAI2300C/file.txt', 'r') as file:
    data = file.read()
print(data)
```

## Question 3: Apply the Tokenization Process to the Text Corpus and Print the First 20 Tokens


```python
# Apply the tokenization process to the text corpus and print the first 20 tokens

def get_tokens(text):
    words = word_tokenize(text)
    return words
tokens = get_tokens(data)

# Print all tokens
print("All Tokens:", tokens)

# Print the first 20 tokens
print("First 20 Tokens:", tokens[:20])
```

## Question 4: Apply Spelling Correction on Each Token and Print the Initial 20 Corrected Tokens as Well as the Corrected Text Corpus


```python
# Apply spelling correction on each token and print the initial 20 corrected tokens as well as the corrected text corpus

# Initialize the Speller object
spell = Speller(lang='en')
spell('lenguage, knowldge')
```

```python
sentence = word_tokenize(data)
print(sentence[:20])
```

```python
# Function to correct spelling in tokens
def correct_spelling(tokens):
    sentence_corrected = ' '.join([spell(word) \
                                   for word in tokens])
    return sentence_corrected

# Correct spelling for the entire sentence
corrected_text = correct_spelling(sentence)

# Tokenize the corrected text to get individual corrected tokens
corrected_tokens = word_tokenize(corrected_text)

# Print the initial 20 corrected tokens
print(corrected_tokens[:20])

# Print the corrected text corpus
print(corrected_text)
```



## Question 5: Apply PoS Tags to Each of the Corrected Tokens and Print them


```python
# Apply PoS tags to each of the corrected tokens and print them

def get_pos(words):
    return pos_tag(words)
get_pos(corrected_tokens)

print(get_pos(corrected_tokens))
```

## Question 6: Remove Stop Words from the Corrected Token List and Print the Initial 20 Tokens


```python
# Remove stop words from the corrected token list and print the initial 20 tokens

stop_words = stopwords.words('english')
print(stop_words)
```

```python
# Print the initial 20 tokens
print(stop_words[:20])
```

```python
def remove_stop_words(sentence_words, stop_words):
    return ' '.join([word for word in sentence_words if \
                     word not in stop_words])
print(remove_stop_words(corrected_tokens, stop_words))
```

## Question 7: Apply Stemming and Lemmatization to the Corrected Token List and then Print the Initial 20 Tokens


```python
# Apply stemming to the corrected token list and then print the initial 20 tokens

def get_stems(word,stemmer):
    return stemmer.stem(word)
porterStem = stem.PorterStemmer()

stemmed_tokens = [get_stems(word, porterStem) for word in corrected_tokens]
# Print the initial 20 tokens after stemming
print(stemmed_tokens[:20])
```

```python
# Apply lemmatization to the corrected token list and then print the initial 20 tokens
lemmatizer = WordNetLemmatizer()
```

```python
def get_lemma(word):
    return lemmatizer.lemmatize(word)

lemmatized_tokens = [get_lemma(word) for word in corrected_tokens]

# Print the initial 20 tokens after stemming and lemmatization
print(lemmatized_tokens[:20])
```

## Question 8: Detect the Sentence Boundaries in the Given Text Corpus and Print the Total Number of Sentences

```python
# Detect the sentence boundaries in the given text corupus and print the total number of sentences

def get_sentences(text):
    return sent_tokenize(text)
get_sentences(data)
```

```python
# Print total number of sentences
print(len(get_sentences(data)))
```
